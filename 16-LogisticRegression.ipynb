```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("Machine_Learning_Dataset/6-bank_customers.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("subscribed", axis=1)
y = df["subscribed"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=15)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score = accuracy_score(y_pred, y_test)
print("accuracy_score: ", score)
print(classification_report(y_pred, y_test))
print("cofusion matrix: \n", confusion_matrix(y_pred, y_test))
```

```python
# hyperparamater tuning
```

```python
model = LogisticRegression()
```

```python
penalty = ["l1","l2","elasticnet"] # cezalari deniyor.
c_values = [100, 10, 1, 0.1, 0.01] # verecegimiz cezanin alpha oranlari 
solver = ["newton-cg","lbfgs","liblinear","sag","saga","newton-cholesky"] # optimize ederken kullanilan algoritmalar
# bunlar bizim deneyecegimiz parametreler bizim bunlari dictionary(sözlüge) cevirmemiz lazim
# gridsearch icin yapiyoruz.
```

```python
params =dict(penalty=penalty, C=c_values, solver=solver)
```

```python
params
```

```python
# grid search cv
```

```python
from sklearn.model_selection import GridSearchCV , StratifiedKFold
```

```python
cv =StratifiedKFold()
```

```python
grid = GridSearchCV(estimator=model, param_grid=params, cv=cv, scoring="accuracy", n_jobs=-1)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_ # bunlar vermemiz gereken en iyi degerler oluyor.
```

```python
grid.best_score_ # bu da elde edilen en iyi skorumuz.
```

```python
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_pred, y_test)
print("accuracy_score: ", score)
print(classification_report(y_pred, y_test))
print("cofusion matrix: \n", confusion_matrix(y_pred, y_test))
```
