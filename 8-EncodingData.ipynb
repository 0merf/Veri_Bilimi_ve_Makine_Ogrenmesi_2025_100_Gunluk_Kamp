```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder , OrdinalEncoder
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
df[["sex","class","embark_town"]]
```

```python
df = df.dropna(subset=["embark_town"]) # embark_town un bos olan iki degerini attik
```

```python
df["sex"].value_counts()
```

```python
df["class"].value_counts()
```

```python
df["embark_town"].value_counts()
```

```python
df_onehot = pd.get_dummies(df,columns=["sex","embark_town"], drop_first=True) 
# bruada sex i ve embark_town u one hot encoding yapiyor sonra oncekileri yani sex ve embark_town u siliyor 
# bu silinenlerin yerine yeni olusan sutunlar geliyor asagida görundugu gibi
```

```python
# LabelEncoder
```

```python
label_encoder =LabelEncoder() # nesne olusturuyoruz.
```

```python
df_label = df.copy()
```

```python
df_label
```

```python
df_label["sex"] = label_encoder.fit_transform(df_label["sex"])
```

```python
df_label
```

```python
df_label["embark_town"] = label_encoder.fit_transform(df_label["embark_town"])
```

```python
df_label
```

```python
# ordinal encoder 
```

```python
df_ordinal = df.copy()
```

```python
class_order = ["Third","Second","First"]
```

```python
ordinal_encoder = OrdinalEncoder(categories=[class_order])
```

```python
df_ordinal["class"] = ordinal_encoder.fit_transform(df_ordinal[["class"]])
```
