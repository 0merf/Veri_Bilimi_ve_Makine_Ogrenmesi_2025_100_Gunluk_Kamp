```python
# merge 
df_merge1 = pd.read_csv("7-merge_data1.csv")
```

```python
df_merge1
```

```python
df_merge2 = pd.read_csv("7-merge_data2.csv")
```

```python
df_merge2
```

```python
# merge - outer join 
df_merge_outer = pd.merge(df_merge1 , df_merge2, on= "Employee_ID", how= "outer") 
# iki tabloyu Employe_ID ye göre birlestiriyor olmayan degerlere null yaziyor.
```

```python
df_merge_outer
```

```python
# merge - left join
df_merge_left = pd.merge(df_merge1 , df_merge2, on= "Employee_ID", how= "left") 
# burada sadece 1. tabloyu baz aliyor ve diger tablolarda 1. tablodaki elemanlarin bilgilerini iceren kisimlar 
# varsa bu bilgileri giriyor yoksa da o kisimlara null atiyor. herseyi 1. tabloya göre yapiyor kisacasi
```

```python
df_merge_left
```

```python
# merge - right join
df_merge_right = pd.merge(df_merge1 , df_merge2, on= "Employee_ID", how= "right") 
# bu da yukaridaki left teki islemi sagdan yapiyor yani ikinci tabloya göre aliyor.
```

```python
df_merge_right
```

```python
# merge - inner join
df_merge_inner = pd.merge(df_merge1 , df_merge2, on= "Employee_ID", how= "inner") 
# bu da iki tablodaki kesisen degerleri alir birlestirir.
```

```python
df_merge_inner
```

```python
# apply 
df = pd.read_csv("8-apply_function_data.csv")
```

```python
def salary_category(salary):
    if salary < 50000:
        return "Low"
    elif 50000 <= salary < 80000:
        return "Medium"
    else:
        return "High"
```

```python
df["Salary"].apply(salary_category) 
# burada salary_category fonksiyonunu df deki Salary stununa uyluluyoruz. 
```

```python
df["Salary_Category"] = df["Salary"].apply(salary_category)
# burada salary_category fonksiyonunu df deki Salary stununa uyluluyoruz. 
# df in icine stun(column) olarak ekliyoruz.
```

```python
df
```

```python
def new_performance_score(row):
    if row["Experience"] > 10:
        return row["Performance_Score"] +1
    else:
        return row["Performance_Score"]
    
```

```python
df["New_Performance_Score"] = df.apply(new_performance_score, axis=1) # axis belirt yoksa hata veriyor
```

```python
df["Formatted_Name"] =df["Name"].apply(lambda x : x.replace("_", " "))
```
