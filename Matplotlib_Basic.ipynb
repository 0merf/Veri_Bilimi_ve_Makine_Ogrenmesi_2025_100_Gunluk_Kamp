```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# bu ifade olusturulan grafiklerin buradan g√∂steririlmesini saglar.
```

```python
age_list = [10,20,30,30,30,40,50,60,70,75]
weight_list = [20,60,80,85,86,87,70,90,95,90]
```

```python
plt.plot(age_list , weight_list, "b")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
# plt.savefig(test.png) grafigi kaydetmek istiyorsak bu sekilde isimlendirerek kaydedebiliriz.
```

```python
numpy_array1 = np.linspace(0,10,20) # 0'dan 10'a kadar esit araliklarla 20 tane sayi uretir.
```

```python
numpy_array1
```

```python
numpy_array2 = numpy_array1 ** 3
```

```python
numpy_array2
```

```python
plt.plot(numpy_array1, numpy_array2)
plt.show()
```

```python
plt.subplot(1,2,1) # ilk kac row olacagi 2. kac stun(column) olacagi 3. de hangi grafigi yazacagini belirtir.
plt.plot(numpy_array1, numpy_array2 , "b*-")
plt.subplot(1,2,2)
plt.plot(numpy_array2, numpy_array1 , "r--")
plt.show()
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.3,0.3])
# 0.1 sagdan ve soldan bosluklari ifade ediyor. 0.3 ler ise grafigin boyutunu ifade ediyor.
# bu degerleri degisirsen grafigin boyu veya olan diger grafikler ile yakinligi uzakligi degisir.
figure_axes.plot(numpy_array1, numpy_array2, "b")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_fig = plt.figure(dpi=100)

new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])

new_axes.plot(numpy_array1, numpy_array1 ** 2, label ="Numpy array ** 2")
new_axes.plot(numpy_array1, numpy_array1 ** 3, label ="Numpy array ** 3")
new_axes.legend() # labelleri grafikte g√∂stermek icin yaziyoruz.
plt.show()
```

```python
# Veri √úretme (Ortalamasƒ± 100, standart sapmasƒ± 15 olan 1000 sayƒ±)
data = np.random.normal(100, 15, 1000)

fig, ax = plt.subplots()
# bins: veriyi ka√ß s√ºtuna b√∂leceƒüimizi belirler
ax.hist(data, bins=30, color='skyblue', edgecolor='black')

ax.set_title("Histogram: Veri Daƒüƒ±lƒ±mƒ±")
ax.set_xlabel("Deƒüerler")
ax.set_ylabel("Frekans (Adet)")
plt.show()
```

```python
# Veri √úretme
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(80, 30, 200) # Daha yayvan
data3 = np.random.normal(90, 20, 200)
data_to_plot = [data1, data2, data3]

fig, ax = plt.subplots()
ax.boxplot(data_to_plot)

ax.set_title("Box Plot: Aykƒ±rƒ± Deƒüer Tespiti")
ax.set_xticklabels(['Grup 1', 'Grup 2', 'Grup 3'])
ax.set_ylabel("Deƒüerler")
plt.show()
# Not: Kutunun dƒ±≈üƒ±ndaki k√º√ß√ºk daireler aykƒ±rƒ± deƒüerlerdir.
```

```python
# Veri √úretme
x = np.random.rand(100) * 100 # 0-100 arasƒ± rastgele deƒüerler
# y, x'e baƒülƒ± olsun ama biraz da rastgelelik (g√ºr√ºlt√º) ekleyelim
y = 2 * x + np.random.normal(0, 10, 100)

fig, ax = plt.subplots()
# alpha: noktalarƒ±n saydamlƒ±ƒüƒ± (√ºst √ºste binenleri g√∂rmek i√ßin)
ax.scatter(x, y, color='purple', alpha=0.6)

ax.set_title("Scatter Plot: ƒ∞ki Deƒüi≈üken Arasƒ±ndaki ƒ∞li≈üki")
ax.set_xlabel("X Deƒüi≈ükeni (√ñrn: Reklam Harcamasƒ±)")
ax.set_ylabel("Y Deƒüi≈ükeni (√ñrn: Satƒ±≈ülar)")
plt.show()
```

```python
# Veri √úretme
epochs = np.arange(1, 51)
# Hata oranƒ± zamanla d√º≈üer (logaritmik bir d√º≈ü√º≈ü sim√ºlasyonu)
loss = np.exp(-epochs / 10) + np.random.normal(0, 0.02, 50)

fig, ax = plt.subplots()
ax.plot(epochs, loss, color='red', marker='o', markersize=4)

ax.set_title("Line Plot: Model Eƒüitim Kaybƒ± (Loss)")
ax.set_xlabel("Epoch (D√∂ng√º Sayƒ±sƒ±)")
ax.set_ylabel("Loss (Hata Oranƒ±)")
ax.grid(True) # Arka plana ƒ±zgara ekler
plt.show()
```

```python
# Veri
kategoriler = ['Elma', 'Armut', 'Muz', '√áilek']
adetler = [150, 90, 200, 120]

fig, ax = plt.subplots()
ax.bar(kategoriler, adetler, color=['red', 'green', 'yellow', 'pink'])

ax.set_title("Bar Chart: Kategori Kar≈üƒ±la≈ütƒ±rmasƒ±")
ax.set_xlabel("Meyve T√ºr√º")
ax.set_ylabel("Adet")
plt.show()
```

```python
# Veri √úretme (Sim√ºle edilmi≈ü bir korelasyon matrisi)
# 1'e yakƒ±n deƒüerler y√ºksek pozitif ili≈üki, -1'e yakƒ±nlar negatif ili≈üki.
correlation_matrix = np.array([
    [1.0, 0.8, 0.1, -0.2],
    [0.8, 1.0, 0.3, -0.5],
    [0.1, 0.3, 1.0, 0.9],
    [-0.2, -0.5, 0.9, 1.0]
])

fig, ax = plt.subplots()
# cmap='coolwarm' : mavi negatif, kƒ±rmƒ±zƒ± pozitif ili≈ükiyi g√∂sterir
cax = ax.imshow(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1)

ax.set_title("Heatmap: Korelasyon Matrisi √ñrneƒüi")
# Renk skalasƒ±nƒ± (colorbar) yanƒ±na ekle
fig.colorbar(cax)

# Eksen etiketlerini ayarla
etiketler = ['√ñzellik A', '√ñzellik B', '√ñzellik C', '√ñzellik D']
ax.set_xticks(np.arange(len(etiketler)))
ax.set_yticks(np.arange(len(etiketler)))
ax.set_xticklabels(etiketler)
ax.set_yticklabels(etiketler)

plt.show()
```
