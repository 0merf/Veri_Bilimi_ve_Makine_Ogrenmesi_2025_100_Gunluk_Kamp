```python
# Multi index
first_index = ["Ogrenci","Ogrenci","Ogrenci","Ogretmen","Ogretmen","Ogretmen"]
inner_index = ["Omer","Ayse","Mustafa","Atil","Samancioglu","Mehmet"]
```

```python
zipped_index = list(zip(first_index, inner_index))
```

```python
zipped_index
```

```python
zipped_index = pd.MultiIndex.from_tuples(zipped_index)
```

```python
zipped_index
```

```python
sample_values = np.ones((6,2))
```

```python
big_df = pd.DataFrame(sample_values, index=zipped_index , columns= ["Age","Note"])
```
