```python
plt.scatter("Height", "Weight", data = data )
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athlete Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height", y="Weight", data = data)
plt.xlabel("Height of Atheletes")
plt.ylabel("Weight of Atheletes")
plt.title("Atheletes Weight vs Height")
plt.show()
```

```python
sns.scatterplot(
    data=data, 
    x="Height", 
    y="Weight", 
    hue="Sex", 
    style="Season", # Veriyi bu stuna gre ekillendirir (K olimpiyatlar daire, Yaz olimpiyatlar arp iareti gibi).
    size="Age", # Noktalarn bykln yaa gre ayarlar (Yallar daha byk nokta).
    sizes=(20, 200),
    alpha=0.7 # Saydamlk. Noktalar st ste bindiinde younluu grebilmek iin (0 ile 1 aras).
)
plt.show()
```

```python
data["Medal"].unique() # Medal stunudaki degerleri grmek icin kullanilir.
```

```python
sns.set_style("whitegrid")
sns.lineplot(x="Height", y="Weight", data = data, hue="Sex")
plt.xlabel("Height of Atheletes")
plt.ylabel("Weight of Atheletes")
plt.title("Atheletes Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(x="Height", data = data, hue="Sex")
plt.ylabel("Frequancy")
plt.title("Atheletes Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(x="Height", data = data, hue="Sex", kind="kde") # yukaridakinin aynisi sadece
# kind kismi degismis kde yapilmis cizgi seklinde gstermek icin.
plt.ylabel("Frequancy")
plt.title("Atheletes Weight vs Height")
plt.show()
```

```python
sns.catplot(x="Medal",y="Height",col="Season", data = data, hue="Sex") # katagorikal olanlari yapmak icin kullanilir.
plt.show()
```

```python
data.corr(numeric_only = True)
```

```python
sns.countplot(
    data=data, 
    x="Medal", 
    hue="Sex",
    palette="viridis",
    order=["Gold", "Silver", "Bronze"] 
)
plt.show()

# palette="viridis": Renk paletini deitirir (pastel, dark, set2 vb. kullanlabilir).
# order=[...]: Stunlarn hangi srada dizileceini manuel belirlersin (Altn en baa gelsin diye).
```

```python
sns.barplot(
    data=data, 
    x="Sex", 
    y="Weight", 
    estimator="mean", # default zaten mean'dir
    errorbar='sd'     # Standart sapma ubuunu gsterir
)
plt.show()

# y="Weight": Y eksenindeki deerin ortalamasn alp ubuun boyunu ona gre izer.
# errorbar='sd': ubuun tepesindeki o ince siyah izgi. Verinin ne kadar yayldn (standart sapmay) gsterir.
```

```python
# Veri ok bykse sadece belli sporlar seip izdirelim
filtered_data = data[data['Sport'].isin(['Basketball', 'Judo', 'Football'])]

sns.boxplot(
    data=filtered_data, 
    x="Sport", 
    y="Age", 
    hue="Sex"
)
plt.show()
# Bu grafik sana medyan (kutunun ortasndaki izgi), verinin %50'sinin yld aral (kutu) ve
# istisnalar (kutunun dndaki noktalar) verir. Basketbolcularn ya ortalamas ile Judocularnkini kyaslaman salar.
```

```python
sns.histplot(
    data=data, 
    x="Age", 
    bins=20, 
    kde=True, 
    color="purple"
)
plt.show()

# bins=20: Veriyi ka kutuya bleceini belirler. Say artarsa detay artar, azalrsa genel hatlar grnr.
# kde=True: (Kernel Density Estimation) Histogramn zerine yumuak bir eri izer. Dalmn eklini (an erisi vb.) daha net gsterir.
```

```python
# nce sadece saysal stunlar seip korelasyon matrisi oluturmalyz
corr_matrix = data[["Age", "Height", "Weight","Year"]].corr()

sns.heatmap(
    corr_matrix, 
    annot=True, 
    cmap="coolwarm", 
    fmt=".2f"
)
plt.show()
# annot=True: Kutucuklarn iine korelasyon saylarn (rn: 0.85) yazar.
# cmap="coolwarm": Mavi (souk/negatif iliki) ile Krmz (scak/pozitif iliki) arasnda renk verir.
# fmt=".2f": Saylar virglden sonra 2 basamak gsterir.
```
